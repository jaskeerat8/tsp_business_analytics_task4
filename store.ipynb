```python
storesecond = store[store["Ship Mode"] == "Second Class"]
storestand = store[store["Ship Mode"] == "Standard Class"]
storefirst = store[store["Ship Mode"] == "First Class"]
storesame = store[store["Ship Mode"] == "Same Day"]
```

```python
x = ["Second Class", "Standard Class", "First Class", "Same Day"]
y = [len(storesecond.index),len(storestand.index),len(storefirst.index),len(storesame.index)]
plt.plot(x,y,label = "Plot line",color = "#00aaff",markeredgecolor = "#000000",marker = "o")
plt.title("Ship mode")
plt.xlabel("Ship Modes available")
plt.ylabel("Number of items shipped ")
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.bar(x,y,label = "bar graph",color = "#fcba03",edgecolor = "#000000")
plt.title("Ship mode")
plt.xlabel("Ship Modes available")
plt.ylabel("Number of items shipped ")
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.pie(y,labels = x,colors = ["#fcba03","#fc054b","#059dfc","#29f505"],autopct="%1.2f%%")
plt.title("Percentage distribution of ship modes")
plt.xlabel("Ship Modes available")
plt.legend()
plt.grid(True)
plt.show()
```

```python
modeprofit = store.groupby("Ship Mode")["Profit"].sum()
modeprofit
```

```python
plt.plot(modeprofit,marker = "o",color = "#093375",label = "Plot line",markeredgecolor = "#000000")
plt.title("Ship Mode versus profit")
plt.xlabel("Ship Modes")
plt.ylabel("Total Profit")
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.pie(modeprofit,labels = ["First Class","Same Day","Second Class","Standard Class"],autopct="%1.2f%%")
plt.title("Percentage of profit by different modes")
plt.xlabel("Different types of Ship Modes")
plt.legend(loc = "lower right")
plt.grid(True)
plt.show()
```

```python
modeprofit = store.groupby("Ship Mode")["Profit"].sum()
segmentprofit = store.groupby("Segment")["Profit"].sum()
regionprofit = store.groupby("Region")["Profit"].sum()
categoryprofit = store.groupby("Category")["Profit"].sum()
```

```python
plt.plot(modeprofit, color = "#ff4d00", lw = 5, ls = "-.", marker = "o",label = "Different Modes",markeredgecolor = "#000000")
plt.plot(segmentprofit, color = "#00b7ff", lw = 5, ls = "-", marker = "x", label ="Different Segments",markeredgecolor = "#000000")
plt.plot(regionprofit, color = "#26ff00", lw = 5, ls = ":", marker = "v", label = "Different Regions",markeredgecolor = "#000000")
plt.plot(categoryprofit, color = "#fffb00", lw = 5, marker = "^", label = "Different Category",markeredgecolor = "#000000")
plt.title("Profit for different Sections")
plt.xticks(rotation = "vertical")
plt.xlabel("Different Sections")
plt.ylabel("Total Profit")
plt.legend()
plt.grid(True)
plt.show()
```

```python
store["Sub-Category"].unique()
```

```python
profitsub = pd.DataFrame(store.groupby("Sub-Category")["Profit"].sum())

plt.plot(profitsub.index,profitsub["Profit"],color = "#000000",lw = 5,markeredgecolor = "#000000")
plt.scatter(profitsub.index,profitsub["Profit"],label = "Profit",color = "#ff0000", s = 150)
plt.legend()
plt.title("Profit/loss for different subcategory")
plt.xlabel("Different Sub-Categories")
plt.ylabel("Total Profit/loss")
plt.xticks(rotation = "vertical")
plt.grid(True)
plt.show()
```

```python
citysalpro = store.groupby("City")[["Sales","Profit"]].sum().sort_values(by = "Sales", ascending = False).head(20)

citysalpro.plot(kind = "bar",edgecolor = "#000000")
plt.title("Profit/loss and Sales of top 20 Cities")
plt.xlabel("Top 20 cities")
plt.ylabel("Total profit / loss and Sales")
plt.grid(True)
plt.show()
```

```python
statesalpro = store.groupby("State")[["Sales","Profit"]].sum().sort_values(by = "Sales", ascending = False).head(20)

statesalpro.plot(kind = "bar",color = ["#ffee00","#0af062"],edgecolor = "#000000")
plt.title("Profit/loss and Sales of top 20 States")
plt.xlabel("States")
plt.ylabel("Total profit / loss and Sales")
plt.grid(True)
plt.show()
```

```python
subquant = store.groupby("Sub-Category")["Quantity"].sum().sort_values(ascending = True).to_frame()

plt.bar(subquant.index,subquant["Quantity"],color = "#ffc400",label = "Quantity",edgecolor = "#000000")
plt.title("Quantity of products sold in ascending order")
plt.xlabel("Product")
plt.ylabel("Total Quantity sold")
plt.legend()
plt.xticks(rotation = "vertical")
plt.grid(True)
plt.show()
```

```python
plt.hist(store["Profit"],bins=(np.arange(0,600,15)),edgecolor = "#000000")
plt.xlabel("Quantity")
plt.ylabel("Profit")
plt.title("Histogram for profit")
```

```python
plt.hist(store["Sales"],bins=(np.arange(0,600,10)),color ="#5eeb34",edgecolor = "#000000")
plt.xlabel("Quantity")
plt.ylabel("Sales")
plt.title("Histogram for Sales")
```
